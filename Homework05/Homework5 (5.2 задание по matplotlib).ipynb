```python
#Exercice3
import numpy as np
import matplotlib.pyplot as plt

# Задача 1: Создание цветового графика sin(x) * sin(y) на интервале [-π, π] × [-π, π]

# Создание сетки значений x и y на интервале [-π, π]
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)

# Вычисление функции sin(x) * sin(y)
Z = np.sin(X) * np.sin(Y)

# Создание цветового графика
plt.figure(figsize=(10, 6))
plt.imshow(Z, cmap='viridis', extent=[-np.pi, np.pi, -np.pi, np.pi])
plt.colorbar(label='sin(x) * sin(y)')
plt.title('Color Plot of sin(x) * sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Задача 2: Создание цветового графика e^(-x^2 - y^2) на интервале [-3, 3] × [-3, 3]

# Создание сетки значений x и y на интервале [-3, 3]
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Вычисление функции e^(-x^2 - y^2)
Z = np.exp(-X**2 - Y**2)

# Создание цветового графика
plt.figure(figsize=(8, 6))
plt.imshow(Z, cmap='plasma', extent=[-3, 3, -3, 3])
plt.colorbar(label='e^(-x^2 - y^2)')
plt.title('Color Plot of e^(-x^2 - y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#exercice1
#задача 2
# Создаем данные
x = np.linspace(5, 25, 250)
y1 = 1 / (1 + x**2)
y2 = 1 / x**2
relative_difference = np.abs(1 - y1 / y2)

# Создаем semilogy график
plt.semilogy(x, relative_difference)
plt.xlabel('x')
plt.ylabel('Относительная разница')
plt.title('Semilogy график относительной разницы')
plt.grid(True)

# Отображаем график
plt.show()


```

```python
#exercice1
#задача 3
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma, psi, erf

x = np.linspace(-5, 5, 100)  # Create an array of x values from -5 to 5

# Compute the values of gamma, psi, and erf functions for each x value
gamma_vals = gamma(x)
psi_vals = psi(x)
erf_vals = erf(x)

# Plot the graphs for each function
plt.figure(figsize=(10, 7))

plt.plot(x, gamma_vals, label='Gamma')
plt.plot(x, psi_vals, label='Psi')
plt.plot(x, erf_vals, label='Erf')

plt.title('Special Functions: Gamma, Psi, and Erf')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()

plt.tight_layout()
plt.show()
```
