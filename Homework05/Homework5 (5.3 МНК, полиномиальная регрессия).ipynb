```python
# 1
#  (n=5)
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)
    w = np.linalg.solve(a, b)  #   
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #  
    return res


x = np.linspace(1, 15, 100)
plt.grid()
plt.title('  :')
plt.plot(x, f(x), color='black')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()

plt.grid()
plt.title(f' ,      n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()

plt.grid()
plt.title(f'    n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='blue')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()
```

```python
#exercice2
#     (n=5)

import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots) + np.random.normal(0, 0.01)
    w = np.linalg.solve(a, b)  #   
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #    
    return res

x = np.linspace(1, 15, 100)
plt.grid()
plt.title('  :')
plt.plot(x, f(x), color='black')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()

plt.grid()
plt.title(f' ,      n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()

plt.grid()
plt.title(f'      n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='blue')
plt.xlabel(': OX')
plt.ylabel(': OY')
plt.show()
```
