```python
#Задание 1
#Полиномиальная регрессия (n=5)
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)
    w = np.linalg.solve(a, b)  #точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение полиномиальной регрессии
    return res


x = np.linspace(1, 15, 100)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен при n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График полиномиальной аппроксимации при n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='blue')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

```python
#exercice2
#Полиномиальная регрессия с зашумлённым данными (n=5)

import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots) + np.random.normal(0, 0.01)
    w = np.linalg.solve(a, b)  #точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение полиномиальной регрессии с шумом
    return res

x = np.linspace(1, 15, 100)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен при n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График полиномиальной аппроксимации с шумом при n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='blue')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

Для решения данной задачи (2) используйте готовую функцию np.linalg.lstsq(), которая аналогично функции  np.linalg.solve() принимает два аргумента: матрицу коэффициентов    и вектор правых частей  . Данная функция возвращает несколько объектов, первый из которых есть решение задачи.
Возьмите   точек на отрезке   и посчитайте в них значение функции  , добавьте гауссов с большим шумом с  . Постройте методом наименьших квадратов кубический полином по данным точкам.

```python
N = 100
n = 4
dots = np.linspace(1, 15, N)
a = np.array([[x**j for j in range(n)] for x in dots])
b = f(dots) + np.random.normal(0, 0.1)

w, resid, rank, s = numpy.linalg.lstsq(a, b)  #точки в полиноме
res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение МНК

x = np.linspace(1, 15, 100)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен n = {n}:')
plt.plot(dots, b, '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График метода наименьших квадратов при n = {n}:')
plt.plot(dots, res, color='blue')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```
